```python
#  create a hashmap
# can do CRUD operations


class Hashmap:
  def __init__(self):
    # creating a list considered to be a ram where datas are allocated
    self.arr=1000
    self.ram=[None for i in range(0,self.arr)]
    print(self.ram)

  def ___hashfun__(self,data):
    # This hash fuction is common for both __add__ and __delete__ so created it as function and calling it whenever it is needed.
    total=0
    for i in data:
      total+=ord(i)
    index=total%10
    return index


  def __add__(self,data):
    # convert the input string into sum of ascci values .
    # apply mod of 10
    #  store the data in the index of the mod result.
    index=self.___hashfun__(data)
    self.ram[index]=data  
    print(self.ram) 

  def __delete__(self,data):
    index=self.___hashfun__(data)
    self.ram[index]=None 
    print(self.ram)


t=Hashmap()
print(t.__add__("ajay"))
print(t.__add__("s1mple"))
print(t.__add__("prayforme"))
print(t.__delete__("s1mple"))

    

    

```
